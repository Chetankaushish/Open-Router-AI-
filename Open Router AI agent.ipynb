```python
# openrouter_api_key is a variable that stores the API key for accessing the OpenRouter API. 
# This key is used to authenticate requests made to the API and allows you to access its features and services.
# Make sure to keep this key secure and do not share it publicly, as it can be used to make unauthorized requests to the API.

openrouter_api_key = "Your_OpenRouter_API_Key_Here"
```

```python
# Import the OpenAI class from the openai package to interact with the OpenRouter API.
# The OpenAI class provides methods to send requests to the API and receive responses, allowing you to utilize the features of the OpenRouter API in your Python code.
from openai import OpenAI

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=openrouter_api_key
)
```

```python
# Use the chat.completions.create method to generate a response from the model.
# You can specify the model you want to use and the messages you want to send to the model.
# In this example, we are sending a message asking "What is the capital of France?" and printing the response from the model.
# You can change the model and the message to test different responses from the model.

response = client.chat.completions.create(
    #model= "nvidia/nemotron-3-nano-omni-30b-a3b-reasoning:free",
    #model= "inclusionai/ling-2.6-1t:free",
    model= "tencent/hy3-preview:free",
    
    messages=[
        {"role": "user", "content": "What is the capital of France?"}
    ]
)
print(response.choices[0].message.content)
```

```python
# You can also ask other questions to the model by changing the content of the message. 
# For example, you can ask "What is the uses of AI?" and see how the model responds.

response = client.chat.completions.create(
    model= "nvidia/nemotron-3-nano-omni-30b-a3b-reasoning:free",
    #model= "inclusionai/ling-2.6-1t:free",
    #model= "tencent/hy3-preview:free",
    
    messages=[
        {"role": "user", "content": "What is the uses of AI?"}
    ]
)
print(response.choices[0].message.content)
```
